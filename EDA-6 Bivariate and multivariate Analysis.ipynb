```python
keys=visa_df['continent'].value_counts().keys()
keys
```

```python
keys=visa_df['continent'].value_counts().keys()
for i in keys:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    count=len(visa_df[con1&con2])
    print("The number of certified applicants from {} is: {}".format(i,count))
    
for i in keys:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Denied'
    count=len(visa_df[con1&con2])
    print("The number of Denied applicants from {} is: {}".format(i,count))
    
```

```python
certified=[]
denied=[]
keys=visa_df['continent'].value_counts().keys()
for i in keys:
    cond1=visa_df['continent']==i
    cond2=visa_df['case_status']=='Certified'
    cond3=visa_df['case_status']=='Denied'
    count1=len(visa_df[cond1&cond2])
    count2=len(visa_df[cond1&cond3])
    certified.append(count1)
    denied.append(count2)
    
pd.DataFrame(zip(certified,denied),columns=['Certified','Denied'],index=keys)
```
