**Method-2 np.where**

- np.where() is applicable for binary conditions

- which means it is applicable only for two lables

- np.where(condition,Truevalue,False_value)

- For example case_status has two lables

- condition:  <col_values>=='Certified'
    
- True value: Replace all certified values with 0
    
- False value: Replace all denied values with 1

```python
############# read the data ################################
file_path="C:\\Users\\lenovo\\Desktop\\Data Science course\\Data Files\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
###############################
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
#Above code gives cetified=0 and not certified=1
visa_df.head()
```

**Label Encoder**

- One of the submodule: preprocessing
    
- Any sklearn packages we have only 3 steps

- Step-1: Read the package
    
- Step-2: Save the package

```python
############# read the data ################################
file_path="C:\\Users\\lenovo\\Desktop\\Data Science course\\Data Files\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
###############################Label Encoder################
# step-1
from sklearn.preprocessing import LabelEncoder
# step-2
le=LabelEncoder()
#step=3:
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df['continent']=le.fit_transform(visa_df['continent'])
print(visa_df[['continent','case_status']].head(10))
```

**One hot encoder**

- one hot means at a time only one will be ON(1), others will be OFF(0)
- For example case status has two unique lables : Certified and Denied
- It will create two new columns

Case_status Certified Denied
Certified 1 0
Denied 0 1

**Advanantage:**
- One hot encoding gives orthogonality between new created columns
- Orthogonality means perpendicular each other
- Perpendicular each other means 90 degrees phase shift
- 90 degress phase shift means independent each other
- Independent means no relation between the columns
- This is very very important property
- Before developing ML model will avoid the dependency

**Draw back**

- But The draw back in Onehot encoder is increase the columns more
- Imagine an original column has 100 uniques lables
- Then One hot encoder will create 100 new columns
- In order to process more columns it require more memory, more time,more hardware
- This is called curse of dimensionality
